```python
#Classify movies based on Ratings[Excellent, Good and Average]
```

```python
def rate(Rating):
    if Rating>=7.0:
        return "Excellent"
    elif Rating>=6.0:
        return "Good"
    else:
        return "Average"
    
    
```

```python
df["rating_category"]=df["Rating"].apply(rate)
```
